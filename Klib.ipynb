```python
import klib
import pandas as pd
```

```python
# missingval_plot
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/akanz1/klib/main/examples/NFL_DATASET.csv', index_col=None)
df
```

```python
klib.missingval_plot(df)
```

```python
# Data Cleaning
df_cleaned = klib.data_cleaning(df)
```

```python
# Correlation plot
klib.corr_plot(df_cleaned, annot=False)
```

```python
# Correlation plot – Only positive correlation
klib.corr_plot(df_cleaned, split='pos', annot=False)
```

```python
# Correlation plot – Only positive correlation
klib.corr_plot(df_cleaned, target='Touchdown')
```

```python
# Correlation plot – Only positive correlation
klib.corr_plot(df_cleaned, split='neg', annot=False)
```

```python
# Correlations dataframez
klib.corr_mat(df_cleaned)
```

```python
# Seeing the distribution
klib.dist_plot(df_cleaned['Win_Prob'])
```

```python
# Categorical values plot
klib.cat_plot(df, figsize=(50, 15))
```
